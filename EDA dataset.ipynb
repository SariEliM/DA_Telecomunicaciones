```python
#Creamos una funcion para evaluar tipo de datos, porcentaje de nulos y cantidad de valores unicos en el dataset
def resumen(df):
    print(f'data shape: {df.shape}')
    resumen = pd.DataFrame(df.dtypes, columns=['data type'])
    resumen['#missing'] = df.isnull().sum()
    resumen['%missing'] = df.isnull().sum()*100/len(df)
    resumen['#unique'] = df.nunique()
    return resumen
```

Para identificar outliers lo haremos de la misma manera que con el dataset anterior

```python
#Primero, crearemos una funcion para quitar acentos y convertir en mayuscula los valores de una columna

import unicodedata

# Función para eliminar acentos y convertir a mayúsculas
def quitar_acentos_y_mayuscula(texto):
    if isinstance(texto, str):  # Verifica que el valor sea una cadena de texto
        texto_normalizado = unicodedata.normalize('NFKD', texto)
        texto_sin_acento = ''.join([c for c in texto_normalizado if not unicodedata.combining(c)])
        return texto_sin_acento.upper()
    return texto
```

```python
#cruzamos con los datos de poblacion por provincia

#antes, capitalizamos los valores de Provincia y quitamos acentos para poder cruzar con exito 
df_tecnologias_acceso_evol['Provincia']=df_tecnologias_acceso_evol['Provincia'].apply(quitar_acentos_y_mayuscula)

# Actualizar el valor de la columna 'Provincia' cuando sea 'Capital Federal'
df_tecnologias_acceso_evol.loc[df_tecnologias_acceso_evol['Provincia'] == 'CAPITAL FEDERAL', 'Provincia'] = 'CABA'


print(f'Shape antes del cruce',df_tecnologias_acceso_evol.shape)
df_tecnologias_acceso_evol_f = pd.merge(df_tecnologias_acceso_evol, df_poblacion_por_provincia_f, on='Provincia', how='left')
print(f'Shape antes del cruce',df_tecnologias_acceso_evol_f.shape)
df_tecnologias_acceso_evol_f.head()
```

Los valores mayores a 100% no son considerados como errores ya que se trata de accesos(conexiones) a internet, por lo que en 1 hogar, pueden existir mas de un acceso

Como un KPI, vamos a medir el crecimiento en % de accesos trimestre a trimestro por provincia, cada 100 hogares de la siguiente manera:

KPI= ((Porc. acceso trim. actual - Porc. acceso trim. anterior)/ Accesos trim. anterior)*100


```python

# Para eso, primero ordenamos el DataFrame por Provincia, Año y Trimestre
df_cobertura_hogar_evol = df_cobertura_hogar_evol.sort_values(by=['Provincia', 'Año', 'Trimestre'])

# Calculamos la diferencia de accesos por cada 100 hogares por provincia
df_cobertura_hogar_evol['Delta acceso'] = df_cobertura_hogar_evol.groupby('Provincia')['Accesos por cada 100 hogares'].diff() 

# Calculamos el porcentaje de crecimiento
df_cobertura_hogar_evol['% Crecimiento'] = (df_cobertura_hogar_evol['Delta acceso'] / df_cobertura_hogar_evol.groupby('Provincia')['Accesos por cada 100 hogares'].shift(1)) * 100

df_cobertura_hogar_evol.head(10)
```

```python
#creamos una columna Periodo combinando Año y Trimestre para una mejor representación
df_cobertura_hogar_evol['Periodo'] = df_cobertura_hogar_evol['Año'].astype(str) + ' T' + df_cobertura_hogar_evol['Trimestre'].astype(str)
 
```

```python

# Creamos un FacetGrid con múltiples subgráficos por provincia para una mirada rapida 
g = sns.FacetGrid(df_cobertura_hogar_evol, col="Provincia", col_wrap=2, height=3, aspect=1.5, sharey=False)
g.map_dataframe(sns.lineplot, x="Periodo", y="% Crecimiento", marker="o")

# Agregamos una línea horizontal en y=2 en cada subgráfico para identificar rapidamente los trimestres que cumplieron o no el KPI
for ax in g.axes.flat:
    ax.axhline(y=2, color='orange', linestyle='--', linewidth=1)

# Ajustamos la visualización
g.set_titles("{col_name}")
g.set_axis_labels("Periodo", "% Crecimiento")
g.set_xticklabels(rotation=45)
plt.subplots_adjust(top=0.9)
g.figure.suptitle('Porcentaje de crecimiento de accesos por cada 100 hogares por provincia', fontsize=16)
```

KPI= ((Nuevo Acceso - Acceso Actual)/ Acceso Actual)*100

"Acceso actual" se refiere al número de hogares con acceso a Internet en el trimestre actual.
